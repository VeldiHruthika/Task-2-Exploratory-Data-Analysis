# **Import Required Libraries**

```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# **Load the Dataset**

```python
# Load Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Display first 5 rows to check data
df.head()
```

# **Summary Statistics**

```python
# Generating summary statistics(basic stats)
print("Summary Statistics:")
print(df.describe(include='all'))
```

# **Histograms for Numerical Features**

```python
# Plotting histograms to see distributions
df.hist(figsize=(12, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.show()
```

# **Boxplots to Detect Outliers**

```python
# Drawing boxplots to check for outliers in Age and Fare
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(y='Age', data=df, color='orange')
plt.title('Boxplot of Age')

plt.subplot(1,2,2)
sns.boxplot(y='Fare', data=df, color='green')
plt.title('Boxplot of Fare')

plt.show()
```

# **Pairplot to Visualize Feature Relationships**

```python
# Pairplot to see relationships between features with survival info
sns.pairplot(df, hue='Survived', diag_kind='kde', palette='husl')
plt.suptitle('Pairplot of Titanic Dataset', y=1.02)
plt.show()
```

# **Correlation Heatmap**

```python
# Heatmap to check correlation between numeric features
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

# **Plotly Interactive Visualization**

```python
# Plotly interactive plot for Age vs Fare vs Survival
fig = px.scatter(df, x='Age', y='Fare', color='Survived', title='Age vs Fare (Interactive)')
fig.show()
```

# **Countplot for Categorical Features**

```python
# Countplot to check how many survived
sns.countplot(x='Survived', data=df, hue='Survived', palette='pastel', legend=False)
plt.title('Survival Count')
plt.show()

# Countplot to see passenger class distribution
sns.countplot(x='Pclass', hue='Pclass', data=df, palette='Set2', legend=False)
plt.title('Passenger Class Count')
plt.show()
```

# **Violin Plot**

```python
# Violin plot to see Age distribution across class and survival
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df, split=True, palette='Set1')
plt.title('Age Distribution across Pclass and Survival')
plt.show()
```

***Basic Observations / Inferences from EDA:***

**1.** Most passengers were in 3rd class *(around 500 people)*

**2.** Majority of passengers were aged between 20 to 40 years.

**3.** Fare feature has many outliers above 300.

**4.** More passengers did not survive *(Survived = 0)* compared to those who survived.

**5.**Higher class passengers *(1st class)* paid higher fares  seen in correlation heatmap.

**6.** Age distribution shows some outliers above 65.

**7.** From the pairplot, younger passengers slightly had better survival chances.

**8.** Fare and Survival have slight positive correlation  people who paid more had better survival chance.

**9.** SibSp and Parch mostly have values 0 or 1  means most people travelled alone or with few family members.
